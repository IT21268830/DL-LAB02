```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
shape = (100000,1)
lam = 2

exponential_array = np.random.exponential(scale=1/lam, size=shape)
uniform_array = np.random.rand(100000)
normal_array = np.random.normal(loc=0, scale=1, size=100000)

plt.figure(figsize=(10, 6))  # Adjust figure size for better visualization
plt.hist(exponential_array, density=True, bins=50, histtype="step", color="red", label="Exponential")
plt.hist(uniform_array, density=True, bins=50, histtype="step", color="blue", label="Uniform")
plt.hist(normal_array, density=True, bins=50, histtype="step", color="green", label="Normal")
```

```python
plt.xlim(-2, 8)  # Adjust x-axis limits to focus on the relevant range
plt.ylim(0, 1.2)  # Adjust y-axis limits for better visualization
plt.legend(loc="upper right")  # Move legend to upper right
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
shape = (100000,1)
lam = 2

exponential_array = np.random.exponential(scale=1/lam, size=shape)
uniform_array = np.random.rand(100000)
normal_array = np.random.normal(loc=0, scale=1, size=100000)

plt.figure(figsize=(10, 6))  # Adjust figure size for better visualization
plt.hist(exponential_array, density=True, bins=50, histtype="step", color="red", label="Exponential")
plt.hist(uniform_array, density=True, bins=50, histtype="step", color="blue", label="Uniform")
plt.hist(normal_array, density=True, bins=50, histtype="step", color="green", label="Normal")

plt.xlim(-2, 8)  # Adjust x-axis limits to focus on the relevant range
plt.ylim(0, 1.2)  # Adjust y-axis limits for better visualization
plt.legend(loc="upper right")  # Move legend to upper right
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Create the meshgrid
x = np.arange(-5, 5.1, 0.1)
y = np.arange(-5, 5.1, 0.1)
X, Y = np.meshgrid(x, y)

# Calculate Z
Z = X**2 + Y**2

# Create the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z)

# Set axis labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set title
ax.set_title('Z = X^2 + Y^2')

plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np

# Load the Pokemon dataset
pokemon = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

# Select the relevant features
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
data = pokemon[features]

# Calculate Pearson correlation
pearson_corr = data.corr(method='pearson')

# Calculate Spearman rank correlation
spearman_corr = data.corr(method='spearman')

# Create heatmaps
plt.figure(figsize=(10, 6))
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Pearson Correlation Heatmap')

plt.figure(figsize=(10, 6))
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Spearman Rank Correlation Heatmap')

plt.show()
```
